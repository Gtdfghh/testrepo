```python
# ==========================
# 1. IMPORT LIBRARIES
# ==========================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ==========================
# 2. CREATE NEW STOCK DATA
# ==========================

np.random.seed(42)

# Generate dates from 2016 to 2025
dates = pd.date_range(
    start='2016-01-01',
    end='2025-01-01',
    freq='W'
)

# Create stock-like prices
trend = np.linspace(1000, 5000, len(dates))

seasonal = 200 * np.sin(
    np.linspace(0, 20, len(dates))
)

noise = np.random.normal(
    0,
    80,
    len(dates)
)

stock_prices = trend + seasonal + noise

# Create DataFrame
df = pd.DataFrame({
    'Date': dates,
    'Close': stock_prices
})

# Set Date as index
df.set_index('Date', inplace=True)

# ==========================
# 3. DISPLAY DATASET
# ==========================

print("FIRST 10 ROWS\n")

print(df.head(10))

print("\nLAST 10 ROWS\n")

print(df.tail(10))

# ==========================
# 4. MAJOR EVENTS
# ==========================

major_events = {
    "2016-06-23": "Brexit Vote",
    "2018-09-04": "Trade War Tensions",
    "2020-03-11": "COVID-19 Pandemic",
    "2021-11-08": "Global Market Recovery",
    "2022-02-24": "Russia-Ukraine War",
    "2023-03-15": "Banking Sector Crisis",
    "2024-06-01": "AI Market Boom"
}

# ==========================
# 5. PLOT ORIGINAL DATA
# ==========================

plt.figure(figsize=(15,6))

plt.plot(
    df['Close'],
    label='Original Stock Prices',
    linewidth=2
)

# Add event lines and labels
for date, event in major_events.items():

    plt.axvline(
        pd.to_datetime(date),
        color='red',
        linestyle='--',
        alpha=0.7
    )

    plt.text(
        pd.to_datetime(date),
        df['Close'].max(),
        event,
        rotation=90,
        fontsize=8
    )

plt.title('Stock Prices with Major Events (2016-2025)')

plt.xlabel('Date')

plt.ylabel('Stock Price')

plt.legend()

plt.show()
```

```python
# ==========================
# 6. TRAIN TEST SPLIT
# ==========================

train_size = int(len(df) * 0.8)

train = df[:train_size]

test = df[train_size:]

print("Training Data Size :", len(train))

print("Testing Data Size :", len(test))

# ==========================
# 7. IMPORT SARIMA LIBRARIES
# ==========================

from statsmodels.tsa.statespace.sarimax import SARIMAX

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

# ==========================
# 8. TRAIN SARIMA MODEL
# ==========================

print("\nTraining SARIMA Model...\n")

model = SARIMAX(
    train['Close'],
    order=(2,1,2),
    seasonal_order=(1,1,1,12)
)

model_fit = model.fit(disp=False)

print("MODEL TRAINED SUCCESSFULLY")

# ==========================
# 9. MAKE PREDICTIONS
# ==========================

predictions = model_fit.predict(
    start=len(train),
    end=len(train)+len(test)-1
)

# ==========================
# 10. EVALUATION METRICS
# ==========================

mae = mean_absolute_error(
    test['Close'],
    predictions
)

rmse = np.sqrt(
    mean_squared_error(
        test['Close'],
        predictions
    )
)

r2 = r2_score(
    test['Close'],
    predictions
)

mape = np.mean(
    np.abs(
        (test['Close'] - predictions)
        / test['Close']
    )
) * 100

# Directional Accuracy

actual_dir = np.sign(
    np.diff(test['Close'])
)

pred_dir = np.sign(
    np.diff(predictions)
)

direction_accuracy = np.mean(
    actual_dir == pred_dir
) * 100

# ==========================
# 11. PRINT RESULTS
# ==========================

print("\n==========================")
print("SARIMA MODEL EVALUATION")
print("==========================")

print(f"MAE : {mae:.2f}")

print(f"RMSE : {rmse:.2f}")

print(f"MAPE : {mape:.2f}%")

print(f"R2 SCORE : {r2:.4f}")

print(f"Directional Accuracy : {direction_accuracy:.2f}%")

print("==========================")

# ==========================
# 12. FINAL PREDICTION GRAPH
# ==========================

plt.figure(figsize=(15,6))

plt.plot(
    test.index,
    test['Close'],
    label='Actual Prices',
    linewidth=2
)

plt.plot(
    test.index,
    predictions,
    label='Predicted Prices',
    linestyle='--',
    linewidth=2
)

plt.title('SARIMA Prediction vs Actual Prices')

plt.xlabel('Date')

plt.ylabel('Stock Price')

plt.legend()

plt.show()
```
